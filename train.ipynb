```python
import joblib
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
from utils.process_weather_data import preprocess_data

raw_df = pd.read_csv('data/weatherAUS.csv')

data = preprocess_data(raw_df)

model = RandomForestClassifier(
    max_leaf_nodes=500,
    random_state=42,
    class_weight='balanced'
)

train_X = data['train_inputs'][data['numeric_cols'] + data['encoded_cols']]
model.fit(train_X, data['train_targets'])

aussie_rain = {
    'model': model,
    'imputer': data['imputer'],
    'scaler': data['scaler'],
    'encoder': data['encoder'],
    'input_cols': data['input_cols'],
    'target_col': data['target_col'],
    'numeric_cols': data['numeric_cols'],
    'categorical_cols': data['categorical_cols'],
    'encoded_cols': data['encoded_cols'],
}

joblib.dump(aussie_rain, 'model/aussie_rain.joblib')
```
